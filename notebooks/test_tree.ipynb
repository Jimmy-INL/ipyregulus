```python
from regulus.utils import io
from regulus.tree.alg import *
from ipyregulus.tree import TreeView 
from ipyregulus import SidePanel
from ipywidgets import interact, interactive, FloatSlider
```

```python
g = io.load('data/gauss4.p')
```

```python
def size(p):
    return p.span[1]-p.span[0] + 2
```

```python
g.tree.data.measures
```

```python
t1 = reduce(g.tree, 
            filter=lambda n: persistence(n) > 0.1,
            select=lambda p: {'id': p.id, 'lvl': p.persistence, 'size': size(p), **p.measures})
```

```python
TreeView(title='fiteness', field='linear_fitness', tree=t1)
```

```python
fp = lambda n, v: persistence(n) > v
tree = g.tree
f = fp
def update(v):
    tv.tree = reduce(tree, 
            filter=lambda n: f(n, v),
            select=lambda p: {'id': p.id, 'lvl': p.persistence, 'size': size(p), **p.measures})
```

```python
sp = SidePanel(title='Tree')
tv = TreeView(title='fiteness', field='linear_fitness', tree=t1)
sl = interactive(update, v=FloatSlider(min=0, max=1, step=0.05))
with sp:
    display(sl, tv)
```
